```python
import os

# Directory containing the files to rename
directory = 'data/val/images'

# Loop through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".png"):
        # Split the filename at '_generated' and keep only the first part
        new_name = filename.split('_generated')[0] + '.png'
        
        # Full path for current file and new file
        old_path = os.path.join(directory, filename)
        new_path = os.path.join(directory, new_name)
        
        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed '{filename}' to '{new_name}'")

print("Renaming complete.")
```
